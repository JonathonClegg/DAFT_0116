```python
import pandas as pd
import numpy as np
df = pd.DataFrame({'TransactionID': [1,2,3,4,5], 'ClientID': [231,765,453,231,892],
                   'Profession': ['Self-employed', 'students', 'Horeca', 'Self-employed', 'finance'], 
                   'Bank_dep': ['009','005', '007', '009', '003'], 
                   'Risk': ['high', 'high', 'medium', 'high', 'low'], 
                   'Number of credits': [2,3,5,2,3], 'Revenue': [30200,12700,89400,30200,740000]})
df
```

Dummy Encoding

```python
pd.get_dummies(df, prefix=['Profession', 'Bank_dep', 'Risk'])
```

Label Encoding

```python
# creating initial dataframe
df_label = pd.DataFrame(df, columns=['Profession', 'Bank_dep', 'Risk'])

# converting type of columns to 'category'
df_label['Profession'] = df_label['Profession'].astype('category')
df_label['Bank_dep'] = df_label['Bank_dep'].astype('category')
df_label['Risk'] = df_label['Risk'].astype('category')

# Assigning numerical values and storing in another column
df_label['Profession_Cat'] = df_label['Profession'].cat.codes
df_label['Bank_dep_Cat'] = df_label['Bank_dep'].cat.codes
df_label['Risk_Cat'] = df_label['Risk'].cat.codes

df["Profession"] = df_label["Profession_Cat"]
df["Bank_dep"] = df_label["Bank_dep_Cat"]
df["Risk"] = df_label["Risk_Cat"]

df
```

Ordinal Encoding


```python
scale_mapper = {"low":1, "medium":2, "high":3}
df["Risk_scale"] = df["Risk"].replace(scale_mapper)
df
```
